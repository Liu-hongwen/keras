```python
import pandas as pd
data = pd.read_csv("data_small.csv")
```

```python
data.shape
```

```python
data.head()
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer = Tokenizer(num_words=90000) # 单词表中最频繁的90000个单词，也就是列数
tokenizer.fit_on_texts(data['words']) # 要用以训练的文本列表，['i love you', 'you love me too']，Series应该也可以
print(tokenizer.word_index) # 词表序号
print(tokenizer.word_counts) # 词频统计
```

```python
binary_matrix = tokenizer.texts_to_matrix(data['words']) # 实际上这是一个binary的向量，如果dictionary的下标为i的那个词在这个string中出现了，那么就给一个1，否则给0
print(binary_matrix) # numpy.ndarray
count_matrix = tokenizer.texts_to_matrix(data['words'], mode='count') # 每个位置代表词频数
print(count_matrix) # numpy.ndarray
freq_matrix = tokenizer.texts_to_matrix(data['words'], mode='freq') # 每个位置代表频率
print(freq_matrix) # numpy.ndarray
tfidf_matrix = tokenizer.texts_to_matrix(data['words'], mode='tfidf') # 每个位置代表tfidf值
print(tfidf_matrix) # numpy.ndarray
```

```python
texts_sequences = tokenizer.texts_to_sequences(data['words']) # 返回各个词序号的列表
print(texts_sequences) # 嵌套的列表
```

```python
from keras.preprocessing.sequence import pad_sequences
texts_sequences_padding = pad_sequences(texts_sequences, maxlen=1000, padding='post', truncating='post') # 短的从后面补0，长的从后面截断
print(texts_sequences_padding) # numpy.ndarray
```
